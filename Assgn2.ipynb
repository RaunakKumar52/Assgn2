```python
#PCA is variance-based. Features with larger scales dominate principal components, making PCA meaningless without standardization. Scaling ensures equal contribution from all features.
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
X = df.drop(columns=["MedHouseVal", "oceanProximity"])
y = df["MedHouseVal"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by PCA")
plt.show()
```

```python
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_scaled)
plt.figure(figsize=(6,6))
plt.scatter(X_pca_2[:,0], X_pca_2[:,1],
            c=y, cmap="viridis", alpha=0.6)
plt.colorbar(label="Median House Value")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Scatter Plot")
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
features = df.drop(columns=["MedHouseVal", "Latitude", "Longitude", "oceanProximity"])
target = df["MedHouseVal"]

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("regressor", LinearRegression())
])

pipeline.fit(X_train, y_train)
```

```python
coef_df = pd.DataFrame({
    "Feature": features.columns,
    "Coefficient": pipeline.named_steps["regressor"].coef_
})

coef_df
pipeline.named_steps["regressor"].intercept_
```

```python
y_pred = pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

n = X_test.shape[0]
p = X_test.shape[1]
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

mse, mae, r2, adj_r2

# no R2 sometimes can indicate overfitting 
# no, low training loss + high test loss does not ensure generalisation
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color="red")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs Actual")
plt.show()
```
